```python
df = pd.read_csv("/content/news_summary.csv", encoding='latin-1', usecols=['headlines', 'text'])
```

```python
df = df.rename(columns={"headlines":"target_text", "text":"source_text"})
df = df[['source_text', 'target_text']]
```

```python
df['source_text'] = "summarize: " + df['source_text']
```

```python
train_df, test_df = train_test_split(df, test_size=0.3)
train_df.shape, test_df.shape
```

```python
from simplet5 import SimpleT5

model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
model.train(train_df=train_df[:500],
            eval_df=test_df[:100], 
            source_max_token_len=128, 
            target_max_token_len=50, 
            batch_size=8, max_epochs=5, use_gpu=True)
```

```python
from flask import Flask, render_template, request ,jsonify
from flask_ngrok import run_with_ngrok
#from image_process import read_image

from flask import Flask, render_template, request, send_file

import time

app = Flask(__name__)
import pickle
model = pickle.load(open('new_model.pkl', 'rb'))

string = ""
text=""

run_with_ngrok(app)
@app.route('/')
def home_page():
    return render_template('index.html', button_name="SUBMIT")

run_with_ngrok(app)
@app.route('/predict', methods=['GET', 'POST'])
def predict():
    if request.method == "POST":
        text = request.form['text']
        size = request.form.get('length')
        #write_to_image(size)

        if size == 'short':
            return render_template('second_page.html', Output=summarization_short(text))
        elif size == 'medium':
           return render_template('second_page.html', Output=summarization(text,4))
        else:
            return render_template('second_page.html', Output=summarization(text,8))

run_with_ngrok(app)
@app.route('/uploader', methods=['GET', 'POST'])
def upload_file():
    if request.method == 'POST':
        f = request.files['file']
        uploaded_file = request.files['file']
        if uploaded_file.filename != '':
            uploaded_file.save(uploaded_file.filename)

    #return render_template('index.html', Output=read_image(uploaded_file.filename))

run_with_ngrok(app)
@app.route('/return-files/')
def return_files_tut():
    try:

        return send_file('press_temp.jpg',
                         attachment_filename='press_release_summary.jpg', as_attachment=True)
    except Exception as e:
        return str(e)

        
        
run_with_ngrok(app)
@app.route('/predict_api', methods=['POST'])
def predict_api():
    '''
    For direct API calls through request
    '''
    data = request.get_json(force=True)
    prediction = summarization(text, 6)
    return jsonify(prediction)




def summarization(input_texxt, n):
    length = len(input_texxt)
    num_words = input_texxt.split()
    length = len(num_words)
    max_chunk = length // n
    input_texxt = input_texxt.replace('.', '.<eos>')
    input_texxt = input_texxt.replace('?', '?<eos>')
    input_texxt = input_texxt.replace('!', '!<eos>')
    sentences = input_texxt.split('<eos>')
    current_chunk = 0
    chunks = []
    for sentence in sentences:
        if len(chunks) == current_chunk + 1:
            if len(chunks[current_chunk]) + len(sentence.split(' ')) <= max_chunk:
                chunks[current_chunk].extend(sentence.split(' '))
            else:
                current_chunk += 1
                chunks.append(sentence.split(' '))
        else:
            print(current_chunk)
            chunks.append(sentence.split(' '))

    for chunk_id in range(len(chunks)):
        chunks[chunk_id] = ' '.join(chunks[chunk_id])
    for i in range(len(chunks)):
        if (i != 0):
            chunks[i] = "summarize:" + chunks[i]
    str = ""
    for i in chunks:
        str += "* " + model.predict(i)[0] + "\n\n"

    return str


run_with_ngrok(app)
app.run()
```
